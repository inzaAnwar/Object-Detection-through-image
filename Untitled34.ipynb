```python
# Step 1: Install YOLOv5
!git clone https://github.com/ultralytics/yolov5
%cd yolov5
!pip install -r requirements.txt

# Step 2: Import libraries
import torch
from IPython.display import display
from PIL import Image as PILImage
import os
import glob

# Step 3: Load YOLOv5x (more accurate than yolov5s)
model = torch.hub.load('ultralytics/yolov5', 'yolov5x', pretrained=True)

# Step 4: Upload images
from google.colab import files
uploaded = files.upload()

# Step 5: Resize + Detect
for filename in uploaded:
    # Resize image
    with PILImage.open(filename) as img:
        resized_img = img.resize((640, 480))  # Smaller image size
        resized_path = f"resized_{filename}"
        resized_img.save(resized_path)

    # Detect objects
    results = model(resized_path)
    results.print()
    results.save()

# Step 6: Show results
exp_folders = sorted(glob.glob('runs/detect/exp*'), key=os.path.getmtime, reverse=True)
latest_folder = exp_folders[0]

for img_path in glob.glob(f'{latest_folder}/*.jpg'):
    print(f"📸 Showing: {img_path}")
    display(PILImage.open(img_path))








```

```python
!pip install ultralytics
from ultralytics import YOLO
from google.colab import files
from PIL import Image
from IPython.display import display
import os
import glob

# Load a more accurate model
model = YOLO("yolov8x.pt")  # More accurate model

# Upload images
uploaded = files.upload()

# Resize + detect
for filename in uploaded:
    img = Image.open(filename)
    img = img.resize((640, 480))
    resized_path = f"resized_{filename}"
    img.save(resized_path)

    # Predict
    results = model(resized_path)
    results[0].save(filename="result.jpg")
    display(Image.open("result.jpg"))
```
